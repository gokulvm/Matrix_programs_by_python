```python
class matrix:

    def addition_of_matrix(self,row,col):
        matrix1 =[[int(input("enter the values for matrix 1: ")) for j in range(col)]for i in range(row)]
        matrix2 =[[int(input("enter the values for matrix 2: ")) for j in range(col)]for i in range(row)]
        #getting the input values from the user
        print("matrix 1")
        for i in range(row):
            for j in range(col):    #print matrix 1
                print(format(matrix1[i][j],"<3"),end=" ")
            print()
        print("----------------")
        print("matrix 2")
        for i in range(row):
            for j in range(col): #print matrix 2
                print(format(matrix2[i][j],"<3"),end=" ")
            print()
        print("----------------") 
        final = [[0 for j in range(col)]for i in range(row)] #creating new zero matrix to save the answer,this have same size of matrix 1 and 2  

        for i in range(row):
            for j in range(col):
                final[i][j] = matrix1[i][j] + matrix2[i][j]  #getting the answers by adding both values

        print("final answer")
        for i in range(row):
            for j in range(col):  #print the answer
                print(format(final[i][j],"<3"),end=" ")
            print()


    def multipication_of_matrix(self,row_p,n,col_q):

        matrix1 = [[int(input("enter the values for matrix 1: ")) for j in range(n)] for i in range(row_p)]
        matrix2 = [[int(input("enter the values for matrix 2: ")) for j in range(col_q)] for i in range(n)]

        final = [[0 for j in range(col_q)] for i in range(row_p)]

        print("Matrix 1")
        for i in range(row_p): #print matrix 1
            for j in range(n):
                print(format(matrix1[i][j],"<3"),end=" ")
            print()


        print("------------------")
        print("Matrix 2")
        for i in range(n):  #print matrix 2
            for j in range(col_q):
                print(format(matrix2[i][j],"<3"),end=" ")
            print()

        print("------------------")

        for i in range(row_p):  #iterating hte matrix 1 rows
            for j in range(col_q):   #iterating the matrix 2 columns
                for k in range(n):   #iterating the matrix 1 column and matrix 2 row   
                    final[i][j] = final[i][j] + matrix1[i][k]*matrix2[k][j]   
        print("Multiplication of matrix 1 and matirx 2") 
        for i in range(row_p):  #print the answer
            for j in range(col_q):
                print(format(final[i][j],"<3"),end=" ")
            print()

    def transpose_of_matrix(self,row,col):
        
        matrix = [[int(input("enter the value : "))for j in range(col)]for i in range(row)]
        final = [[0 for j in range(row)]for i in range(col)]

        for i in range(row):
            for j in range(col):
                final[j][i] = matrix[i][j]

        print("original values")
        for i in range(row):
            for j in range(col):
                print(format(matrix[i][j],"<3"),end=" ")
            print()

        print("----------------------------")

        print("Transpose matrix")
        for i in range(col):
            for j in range(row):
                print(format(final[i][j],"<3"),end=" ")
            print()


ans = matrix()
ans.addition_of_matrix(3,3)
ans.multipication_of_matrix(2,3,4)
ans.transpose_of_matrix(3,4)
```
